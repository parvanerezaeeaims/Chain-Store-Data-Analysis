```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
from apyori import apriori
```

```python
df= pd.read_csv(r"/home/ccp/Desktop/task1/Stores_Transactions .csv")
```

```python
data_temp = df.copy()
data_temp['product_name']=data_temp['product_name']+','
grouped=data_temp.groupby(['user_id', 'transaction_date']).agg({'product_name': 'sum'})
records = []
for line in grouped['product_name']:
    items = line.split(',')
    temp = []
    for item in items:
        if item.strip() != '':
            temp.append(item)
    records.append(temp)
records
```

```python
result = apriori(records, min_support=0.0005, min_confidence=0.007, min_lift=1.1)

for res in result:
    for item in res[2]:
        print('items_base:', list(item.items_base))  
        print('items_add:', list(item.items_add))    
        
        if item.items_base:  # Only print if items_base is not empty
            print(', '.join(list(item.items_base)), '---------->', ', '.join(list(item.items_add)))
        else:
            print('No item base ---------->', ', '.join(list(item.items_add)))
        
        print('support:', res[1])
        print('confidence:', item.confidence)
        print('lift:', item.lift)
        print()
        print('------------------------------------------------------------')
        print()
```

min_support:how frequent an itemset is in the dataset.If you set min_support=0.05, it means the item must appear in at least 5% of all transactions to be considered.
min_confidence It measures how strong the rule is.
min_lift It measures how strong the relationship is between two items.
